```python
# REPLACING TARGET VARIABLE DATA TO 1's & 0's :

data.loc[ data['diagnosis'] == 'M' ,'diagnosis'] = 0   # DISADVANTAGE : data will be of dtype <object> instead of <int>
data.loc[ data['diagnosis'] == 'B' ,'diagnosis'] = 1

data = data.astype({"diagnosis":'int'})                # Converting column to dtype <int>

data.head()
```

```python
# SEPERATING INPUT AND TARGET VARIABLES :

x=data.drop(['diagnosis'],axis=1)
x.head()
x.shape
```

```python
y=data['diagnosis']
y.head()
y.shape
```

<b>IMPLEMENTING VARIOUS ML MODELS</b>

<b> Train-Test Split : </b>

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

x_train, x_test,y_train, y_test =train_test_split(x,y,train_size=0.9,random_state=10)
```

```python
print('x      :', x.shape)
print('y      :', y.shape)
print('x_train:', x_train.shape)
print('y_train:', y_train.shape)
print('x_test:', x_test.shape)
print('y_test:', y_test.shape)
```

```python
print(y.mean(), y_train.mean(), y_test.mean())
```

<b>LOGISTIC REGRESSION :</b>

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()

lr.fit(x_train,y_train)                                 # TRAINING THE MODEL
y_true = lr.predict(x_train)                            # PREDICTION ON TRAINED DATA

accuracy = metrics.accuracy_score(y_true,y_train)
print(f"Accuracy on Train data = {round(accuracy*100,2)} %")          # TRAIN DATA ACCURACY


y_required=lr.predict(x_test)                          # PREDICTION ON TEST DATA

accuracy_lr=metrics.accuracy_score(y_required,y_test)          # TEST DATA ACCURACY
print(f"Accuracy on Test data  = {round(accuracy_lr*100,2)} %")          # TRAIN DATA ACCURACY
 
```

<b>KNN MODEL :</b>

```python
# KNN MODEL :

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# instantiate learning model (k = 3)
knn = KNeighborsClassifier(n_neighbors=3)

# Training the Classifier/fitting the model :
knn.fit(x_train, y_train)

y_true = knn.predict(x_train)                                             # PREDICTION ON TRAINED DATA

accuracy = metrics.accuracy_score(y_true,y_train)
print(f"Accuracy on Train data = {round(accuracy*100,2)} %")              # TRAIN DATA ACCURACY


y_required=knn.predict(x_test)                                            # PREDICTION ON TEST DATA

accuracy_knn=metrics.accuracy_score(y_required,y_test)                    # TEST DATA ACCURACY
print(f"Accuracy on Test data  = {round(accuracy_knn*100,2)} %")          # TRAIN DATA ACCURACY


#input_data = (16.02,23.24,102.7,797.8,0.08206,0.06669,0.03299,0.03323,0.1528,0.05697,0.3795,1.187,2.466,40.51,0.004029,0.009269,0.01101,0.007591,0.0146,0.003042,19.19,33.88,123.8,1150,0.1181,0.1551,0.1459,0.09975,0.2948,0.08452)
input_data=(13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)
# Change the tuple data into numpy_array for faster processing
input_data_as_numpy_array=np.asarray(input_data)

test_data_for_prediction=input_data_as_numpy_array.reshape(1,-1)  # Reshape it,so that 2d array as expected is formed
prediction=knn.predict(test_data_for_prediction)

if prediction==0:
    prediction="Malignant"
else:
    prediction="Benign"
    
print(f"The Breast Cancer State is in {prediction.upper()} stage")
```

<b>SVM MODEL :</b>

```python
# SVM MODEL :

#Import svm model
from sklearn import svm
from sklearn.svm import SVC
from sklearn import metrics

#Create a svm Classifier
svm = SVC(C=1, kernel='rbf')

# Train the Classifier/fitting the model
svm.fit(x_train, y_train)

y_true = svm.predict(x_train)                                             # PREDICTION ON TRAINED DATA

accuracy = metrics.accuracy_score(y_true,y_train)
print(f"Accuracy on Train data = {round(accuracy*100,2)} %")              # TRAIN DATA ACCURACY


y_required=svm.predict(x_test)                                            # PREDICTION ON TEST DATA

accuracy_svm=metrics.accuracy_score(y_required,y_test)                    # TEST DATA ACCURACY
print(f"Accuracy on Test data  = {round(accuracy_svm*100,2)} %")          # TRAIN DATA ACCURACY


#input_data = (16.02,23.24,102.7,797.8,0.08206,0.06669,0.03299,0.03323,0.1528,0.05697,0.3795,1.187,2.466,40.51,0.004029,0.009269,0.01101,0.007591,0.0146,0.003042,19.19,33.88,123.8,1150,0.1181,0.1551,0.1459,0.09975,0.2948,0.08452)
input_data=(13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)
# Change the tuple data into numpy_array for faster processing
input_data_as_numpy_array=np.asarray(input_data)

test_data_for_prediction=input_data_as_numpy_array.reshape(1,-1)  # Reshape it,so that 2d array as expected is formed
prediction=svm.predict(test_data_for_prediction)

if prediction==0:
    prediction="Malignant"
else:
    prediction="Benign"
    
print(f"The Breast Cancer State is in {prediction.upper()} stage")
```

<b>RANDOM FOREST METHOD :</b>

```python
# RANDOM FOREST METHOD :

from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics                                           # TO CHECK ACCURACY OF PREDICTION

# Create a random forest Classifier
rfc = RandomForestClassifier(n_jobs=2, random_state=0)

# Train the Classifier/fitting the model
rfc.fit(x_train,y_train)                                              # TRAINING THE INPUT_DATA & OUTPUT EXPECTED 

y_true = rfc.predict(x_train)                                         # PREDICTION ON TRAINED DATA

accuracy = metrics.accuracy_score(y_true,y_train)
print(f"Accuracy on Train data = {round(accuracy*100,2)} %")          # TRAINED DATA ACCURACY


y_required=rfc.predict(x_test)                                        # PREDICTION ON TEST DATA

accuracy_rfc=metrics.accuracy_score(y_required,y_test)                # TEST DATA ACCURACY
print(f"Accuracy on Test data  = {round(accuracy_rfc*100,2)} %")      # TRAIN DATA ACCURACY


#input_data = (16.02,23.24,102.7,797.8,0.08206,0.06669,0.03299,0.03323,0.1528,0.05697,0.3795,1.187,2.466,40.51,0.004029,0.009269,0.01101,0.007591,0.0146,0.003042,19.19,33.88,123.8,1150,0.1181,0.1551,0.1459,0.09975,0.2948,0.08452)
input_data=(13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# Change the tuple data into numpy_array for faster processing
input_data_as_numpy_array=np.asarray(input_data)

test_data_for_prediction=input_data_as_numpy_array.reshape(1,-1)  # Reshape it,so that 2d array as expected is formed
prediction=rfc.predict(test_data_for_prediction)


if prediction==0:
    prediction="Malignant"
else:
    prediction="Benign"
    
print(f"The Breast Cancer State is in {prediction.upper()} stage")
```
